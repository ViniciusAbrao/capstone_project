# Prediction results:
The first section takes the full list of users and compare the predicted results with the labels.pkl file, computing the accuracy, precision and recall for each offer. Next, it is possible to input the features of one user and predict whether or not someone will respond to an offer.

```python
import numpy as np
import pandas as pd
import pickle
from sklearn.metrics import classification_report
```

```python
#Load files

with open('features.pkl', 'rb') as file:
    X = pickle.load(file)
X=np.array(X)
columns_x=['age','gender','income','became_member_on']
x_df=pd.DataFrame(X,columns=columns_x)

with open('labels.pkl', 'rb') as file:
    Y = pickle.load(file)
Y=np.array(Y)
labels=['y0','y1','y2','y3','y4','y5','y6','y7','y8','y9']
y_df=pd.DataFrame(Y,columns=labels)

df=pd.concat([x_df,y_df],axis=1)
```

```python
#Prediction for all users

features=X

for label in labels:
    print(label)
    name=label+'.pkl'
    with open(name, 'rb') as file:
        y_pred = pickle.load(file).predict(features) 
    y_real=y_df[label].values
    print(classification_report(y_real,y_pred))
```

```python
import numpy as np
import pandas as pd
import pickle
from sklearn.metrics import classification_report
```

```python
with open('features.pkl', 'rb') as file:
    X = pickle.load(file)
X=np.array(X)
columns_x=['age','gender','income','became_member_on']
x_df=pd.DataFrame(X,columns=columns_x)

with open('labels.pkl', 'rb') as file:
    Y = pickle.load(file)
Y=np.array(Y)
labels=['y0','y1','y2','y3','y4','y5','y6','y7','y8','y9']
y_df=pd.DataFrame(Y,columns=labels)

df=pd.concat([x_df,y_df],axis=1)
```

```python
#Prediction for all users

features=X

for label in labels:
    print(label)
    name=label+'.pkl'
    with open(name, 'rb') as file:
        y_pred = pickle.load(file).predict(features) 
    y_real=y_df[label].values
    print(classification_report(y_real,y_pred))
```

```python
'''
Prediction for one user- Some examples:

        age         gender(1=F,2=M,3=O)  income     became_member_on
       [6.8000000e+01, 2.0000000e+00, 7.0000000e+04, 2.0180426e+07]
       [4.9000000e+01, 2.0000000e+00, 7.3000000e+04, 2.0170126e+07]
       [8.3000000e+01, 1.0000000e+00, 5.0000000e+04, 2.0160307e+07]
       [6.2000000e+01, 1.0000000e+00, 8.2000000e+04, 2.0170722e+07]
'''

#Enter the user features in the line below
features=np.array([[6.2000000e+01, 1.0000000e+00, 8.2000000e+04, 2.0170722e+07]])

#Compute and print the predicted and real values
y_pred_resp=[]
for label in labels:
    name=label+'.pkl'
    with open(name, 'rb') as file:
        y_pred = pickle.load(file).predict(features)
    y_pred_resp.append(y_pred)

predicted=(np.array(y_pred_resp)).reshape(-1)
real=df.loc[(df['age']==features[0][0]) & (df['gender']==features[0][1]) & (df['income']==features[0][2]) &\
       (df['became_member_on']==features[0][3])].loc[:,labels].values.reshape(-1)
print("The predicted values are: {}".format(predicted))
print("The real values are     : {}".format(real))
```
